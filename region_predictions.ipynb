```python
import os
import json

import numpy as np
import pandas as pd
import pycm
```

```python
region_predictions_path = "output/region_predictions_aggregated.json"
data = None
with open(region_predictions_path) as f:
    data = json.load(f)

labels_path = "output/dataset_labels.json"
labels = None
with open(labels_path) as f:
    labels = json.load(f)

labels_map = {'Mild':0, 'Moderate':1, 'Severe':2}
```

```python
actual_vector = []
predict_vector = []
for filename in labels.keys():
    actual = labels_map[labels[filename]]
    prediction = int(max(data[filename], key=data[filename].get))
    actual_vector.append(actual)
    predict_vector.append(prediction)
```

```python
cm = pycm.ConfusionMatrix(
    actual_vector = actual_vector,
    predict_vector = predict_vector
)
```

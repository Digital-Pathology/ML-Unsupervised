```python
from feature_manager import FeatureManager
from PIL import Image
import numpy as np
from os import walk, path
from matplotlib.colors import rgb_to_hsv
from pprint import pprint

def load_image(fn):
    img = Image.open(fn)
    img.load()
    data = np.asarray(img, dtype="int32")
    return data
```
